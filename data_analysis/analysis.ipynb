```python
def column_name_to_initials(column_name):
    return ''.join([word[0] for word in column_name.split()]).upper()
```

```python
import numpy as np


# Define the columns to be split
columns_to_split = ['Overall Value', 'Offensive Performance', 'Defensive Performance', 'Biometrics', 'Migration Difficulty']

# Loop through each column, split by ' ' to separate entries, then split by ':' to create subcolumns
for column in columns_to_split:
    # Fill empty strings '' with nan values
    df[column] = df[column].replace('', np.nan)
    # Handle missing values by replacing them with a placeholder that ensures proper splitting
    df[column] = df[column].fillna(': ')
    
    # Split into initial parts
    sub_cols = df[column].str.split(' ', expand=True)
    
    for i in range(sub_cols.shape[1]): 
        # Create names for the sub-columns
        sub_col_name = f'{column_name_to_initials(column)}: {i+1}'
        sub_col_score_name = f'{column_name_to_initials(column)}: {i+1} score'
        
        # Split each part into value and score
        df[[sub_col_name, sub_col_score_name]] = sub_cols[i].str.split(':', expand=True)

        # Handle conversion to float for the score columns and ignore the value columns
        try:
            # df[sub_col_score_name] = df[sub_col_score_name].fillna(0.0)
            df[sub_col_score_name] = pd.to_numeric(df[sub_col_score_name], errors='coerce')
            df[sub_col_score_name] = df[sub_col_score_name].astype(float).round(1)
        except ValueError:
            print(f'Error converting {sub_col_score_name} to float')

    # Compute the average of the score marked subcolumns for each main column
    score_subcolumns = [f'{column_name_to_initials(column)}: {j+1} score' for j in range(sub_cols.shape[1])]
    print("score_subcolumns:", score_subcolumns)
    df[column] = df[score_subcolumns].mean(axis=1, skipna=True).round(1)
    

```

```python
# Drop values not in [0,100]
for column in columns_to_check:
    # df = df[(df[column] >= 0) & (df[column] <= 100)]
    df[column] = np.clip(df[column], 0, 100)
```

```python
# Define the target mean for normalization
target_mean = 50

# Iterate over each group in the dataframe grouped by 'League'
for league, group in df.groupby('League'):
    print(f'Processing league: {league}')
    for column in columns_to_check:
        print(f'\tStandardizing: {column}')

        # Calculate mean and standard deviation of the column for the current league
        mean = group[column].mean()
        std = group[column].std()
        min_ = group[column].min()
        max_ = group[column].max()
        print(f'\t\tMean: {mean:.2f}, Std: {std:.2f}, Min: {min_:.2f}, Max: {max_:.2f}')

        # Calculate scale to bring mean to target_mean while considering the min/max after shifting
        min_after_shift = min_ - mean + target_mean
        max_after_shift = max_ - mean + target_mean

        # Calculate scale factors to avoid exceeding [0, 100]
        scale_down = min(target_mean / (target_mean - min_after_shift), (100 - target_mean) / (max_after_shift - target_mean))

        # Apply scaling and shifting
        group[column] = (group[column] - mean) * scale_down + target_mean

        # Clip values to ensure they remain within [0, 100]
        group[column] = np.clip(group[column], 0, 100).round(1)

        # Update the main dataframe with the normalized data for the current league and column
        df.loc[group.index, column] = group[column]

        # Display the new statistics for the column
        mean = group[column].mean()
        std = group[column].std()
        min_ = group[column].min()
        max_ = group[column].max()
        print(f'\t\tMean: {mean:.2f}, Std: {std:.2f}, Min: {min_:.2f}, Max: {max_:.2f}\n')
```
